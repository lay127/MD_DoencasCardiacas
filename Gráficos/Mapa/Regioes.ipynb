```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Baixando um shapefile dos estados do Brasil (GeoJSON)
url = 'https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson'
estados = gpd.read_file(url)

# Mapeia cada estado para sua regio
regioes = {
    'Acre': 'Norte',
    'Alagoas': 'Nordeste',
    'Amap': 'Norte',
    'Amazonas': 'Norte',
    'Bahia': 'Nordeste',
    'Cear': 'Nordeste',
    'Distrito Federal': 'Centro-Oeste',
    'Esprito Santo': 'Sudeste',
    'Gois': 'Centro-Oeste',
    'Maranho': 'Nordeste',
    'Mato Grosso': 'Centro-Oeste',
    'Mato Grosso do Sul': 'Centro-Oeste',
    'Minas Gerais': 'Sudeste',
    'Par': 'Norte',
    'Paraba': 'Nordeste',
    'Paran': 'Sul',
    'Pernambuco': 'Nordeste',
    'Piau': 'Nordeste',
    'Rio de Janeiro': 'Sudeste',
    'Rio Grande do Norte': 'Nordeste',
    'Rio Grande do Sul': 'Sul',
    'Rondnia': 'Norte',
    'Roraima': 'Norte',
    'Santa Catarina': 'Sul',
    'So Paulo': 'Sudeste',
    'Sergipe': 'Nordeste',
    'Tocantins': 'Norte'
}
estados['regiao'] = estados['name'].map(regioes)

# Define as regies que voc quer destacar
regioes_desejadas = ['Centro-Oeste', 'Nordeste']
estados_destaque = estados[estados['regiao'].isin(regioes_desejadas)]

# Plotando o mapa
fig, ax = plt.subplots(figsize=(10, 10))
estados.plot(ax=ax, color='lightgrey', edgecolor='black')  # fundo do mapa
estados_destaque.plot(ax=ax, color='orange')               # regies pintadas

# Ttulo e ajuste visual
plt.title('Regies Centro-Oeste e Nordeste do Brasil', fontsize=14)
plt.axis('off')
plt.show()
```
