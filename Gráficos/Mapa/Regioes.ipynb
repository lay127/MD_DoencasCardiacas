```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Baixando um shapefile dos estados do Brasil (GeoJSON)
url = 'https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson'
estados = gpd.read_file(url)

# Mapeia cada estado para sua região
regioes = {
    'Acre': 'Norte',
    'Alagoas': 'Nordeste',
    'Amapá': 'Norte',
    'Amazonas': 'Norte',
    'Bahia': 'Nordeste',
    'Ceará': 'Nordeste',
    'Distrito Federal': 'Centro-Oeste',
    'Espírito Santo': 'Sudeste',
    'Goiás': 'Centro-Oeste',
    'Maranhão': 'Nordeste',
    'Mato Grosso': 'Centro-Oeste',
    'Mato Grosso do Sul': 'Centro-Oeste',
    'Minas Gerais': 'Sudeste',
    'Pará': 'Norte',
    'Paraíba': 'Nordeste',
    'Paraná': 'Sul',
    'Pernambuco': 'Nordeste',
    'Piauí': 'Nordeste',
    'Rio de Janeiro': 'Sudeste',
    'Rio Grande do Norte': 'Nordeste',
    'Rio Grande do Sul': 'Sul',
    'Rondônia': 'Norte',
    'Roraima': 'Norte',
    'Santa Catarina': 'Sul',
    'São Paulo': 'Sudeste',
    'Sergipe': 'Nordeste',
    'Tocantins': 'Norte'
}
estados['regiao'] = estados['name'].map(regioes)

# Define as regiões que você quer destacar
regioes_desejadas = ['Centro-Oeste', 'Nordeste']
estados_destaque = estados[estados['regiao'].isin(regioes_desejadas)]

# Plotando o mapa
fig, ax = plt.subplots(figsize=(10, 10))
estados.plot(ax=ax, color='lightgrey', edgecolor='black')  # fundo do mapa
estados_destaque.plot(ax=ax, color='orange')               # regiões pintadas

# Título e ajuste visual
plt.title('Regiões Centro-Oeste e Nordeste do Brasil', fontsize=14)
plt.axis('off')
plt.show()
```
