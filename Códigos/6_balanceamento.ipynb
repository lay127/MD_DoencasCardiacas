## Separação dos valores de treino e teste

```python
X_prev = base.drop(columns=['Q06306'])
y_classe = base['Q06306']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_prev
```

```python
y_classe
```

```python
y_classe.shape
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X_prev, y_classe, test_size=0.3, random_state=42)
```

```python
X_teste.shape
```

```python
y_teste
```

```python
y_treino.value_counts()
```

```python
y_teste.value_counts()
```

```python
import numpy as np
```

```python
np.savez('../Bases/Criados/Treino_Teste/treino_teste.npz', X_treino=X_treino, X_teste=X_teste, y_treino=y_treino, y_teste=y_teste, feature_names=X_treino.columns)
```

```python
df_treino_original = pd.concat([X_treino.reset_index(drop=True), y_treino.reset_index(drop=True)], axis=1)
df_teste_original = pd.concat([X_teste.reset_index(drop=True), y_teste.reset_index(drop=True)], axis=1)

df_treino_original.to_csv("../Bases/Criados/Treino_Teste/treino.csv", index=False)
df_teste_original.to_csv("../Bases/Criados/Treino_Teste/teste.csv", index=False)
```

### OverSampling

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_treino, y_treino)
```

```python
np.savez('../Bases/Criados/Treino_Teste/over.npz', X_treino=X_resampled, X_teste=X_teste, y_treino=y_resampled, y_teste=y_teste, feature_names=X_treino.columns)
```

```python
df_over = pd.concat([pd.DataFrame(X_resampled, columns=X_treino.columns),
                     pd.DataFrame(y_resampled, columns=['Q06306'])], axis=1)
df_over.to_csv("../Bases/Criados/Treino_Teste/over.csv", index=False)
```

### UnderSampling

```python
from imblearn.under_sampling import TomekLinks
balanceamento_under = TomekLinks(sampling_strategy='auto')
X_under, y_under = balanceamento_under.fit_resample(X_treino, y_treino)
```

```python
np.savez('../Bases/Criados/Treino_Teste/under.npz', X_treino=X_under, X_teste=X_teste, y_treino=y_under, y_teste=y_teste, feature_names=X_treino.columns)
```

```python
df_under = pd.concat([pd.DataFrame(X_under, columns=X_treino.columns),
                      pd.DataFrame(y_under, columns=['Q06306'])], axis=1)
df_under.to_csv("../Bases/Criados/Treino_Teste/under.csv", index=False)
```

### Random UnderSampling

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = undersample.fit_resample(X_treino, y_treino)
```

```python
np.savez('../Bases/Criados/Treino_Teste/random.npz', X_treino=X_resampled, X_teste=X_teste, y_treino=y_resampled, y_teste=y_teste, feature_names=X_treino.columns)
```

```python
df_random = pd.concat([pd.DataFrame(X_resampled, columns=X_treino.columns),
                       pd.DataFrame(y_resampled, columns=['Q06306'])], axis=1)
df_random.to_csv("../Bases/Criados/Treino_Teste/random.csv", index=False)
```
